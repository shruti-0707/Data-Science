```python
# Assuming the values are in the first cell of the first row
first_row_values = df.iloc[0, 0].split(',')

# Create a new DataFrame with these values as columns
new_df = pd.DataFrame([first_row_values])
```

```python
# Drop the first row from the original DataFrame
df = df.drop(0)

# Reset the index of the original DataFrame
df.reset_index(drop=True, inplace=True)

# Concatenate the new DataFrame with the original one
final_df = pd.concat([new_df, df], axis=0)# Drop the first row from the original DataFrame
df = df.drop(0)

# Reset the index of the original DataFrame
df.reset_index(drop=True, inplace=True)

# Concatenate the new DataFrame with the original one
final_df = pd.concat([new_df, df], axis=0)
```

```python
final_df.head()
```

```python
df1=pd.get_dummies(final_df)
df1.head()
```

```python
frequent_itemsets = apriori(df1, min_support=0.001, use_colnames=True)#df-dataframe, min_support default=0.5
# min_support is a float between 0 and 1 for minumum support of the itemsets returned.
# The support is computed as the fraction transactions_where_item(s)_occur / total_transactions.
frequent_itemsets # output is frequent itemsets: 1-item itemsets, 2-item itemsets etc. with min support criteria
```

```python
print(frequent_itemsets)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7) # min_threshold is Confidence
rules # http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/
rules.sort_values('lift',ascending = False)[0:20]
```
