# Exercise 1: Prime Numbers

```python
import math

def is_prime(number):
    """Check if a number is a prime number."""
    if number <= 1:
        return False
    if number == 2:
        return True  # 2 is the only even prime number
    if number % 2 == 0:
        return False  # Other even numbers are not prime

    # Check for factors from 3 to sqrt(number)
    limit = int(math.sqrt(number)) + 1
    for i in range(3, limit, 2):
        if number % i == 0:
            return False

    return True

# Example usage
if __name__ == "__main__":
    try:
        num = int(input("Enter a number to check if it's prime: "))
        if is_prime(num):
            print(f"{num} is a prime number.")
        else:
            print(f"{num} is not a prime number.")
    except ValueError:
        print("Please enter a valid integer.")
```

# Exercise 2: Product of Random Numbers

```python
import random

def main():
    # Generate two random numbers between 1 and 10
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)

    # Calculate the correct product
    correct_product = num1 * num2

    # Ask the user to enter their answer
    print(f"Calculate the product of {num1} and {num2}:")

    try:
        user_answer = int(input("Your answer: "))

        # Check if the user's answer is correct
        if user_answer == correct_product:
            print("Correct! Well done.")
        else:
            print(f"Incorrect. The correct answer is {correct_product}.")

    except ValueError:
        print("Please enter a valid integer.")

if __name__ == "__main__":
    main()
```

# Exercise 3: Squares of Even/Odd Numbers

```python
# Python script to print the squares of all even numbers within the range of 100 to 200

# Choice: Even numbers

def print_squares_of_evens(start, end):
    print(f"Squares of even numbers between {start} and {end}:")
    for number in range(start, end + 1):
        if number % 2 == 0:  # Check if the number is even
            square = number ** 2  # Calculate the square of the number
            print(f"{number}^2 = {square}")

# Define the range
start_range = 100
end_range = 200

# Call the function
print_squares_of_evens(start_range, end_range)
```

# Exercise 4: Word counter

```python
def count_words(text):
    # Convert the text to lowercase to ensure case-insensitive counting
    text = text.lower()

    # Initialize an empty dictionary to store word counts
    word_counts = {}

    # Split the text into words by spaces
    words = text.split()

    # Iterate through each word in the list
    for word in words:
        # Remove any punctuation attached to the word
        cleaned_word = ''.join(char for char in word if char.isalnum())

        # Count occurrences of each word
        if cleaned_word in word_counts:
            word_counts[cleaned_word] += 1
        else:
            word_counts[cleaned_word] = 1

    return word_counts

# Example usage
if __name__ == "__main__":
    input_text = "This is a sample text. This text will be used to demonstrate the word counter."
    word_counts = count_words(input_text)

    print("Word counts:")
    for word, count in word_counts.items():
        print(f"'{word}': {count}")
```

# Exercise 5: Check for Palindrome

```python
def is_palindrome(s):
    # Initialize an empty string to build the cleaned version
    cleaned_string = ''

    # Iterate through each character in the input string
    for char in s:
        # Check if the character is alphanumeric
        if char.isalnum():
            # Convert to lowercase and append to cleaned_string
            cleaned_string += char.lower()

    # Check if cleaned_string is equal to its reverse
    return cleaned_string == cleaned_string[::-1]

# Example usage
if __name__ == "__main__":
    test_strings = [
        "A man, a plan, a canal, Panama",
        "racecar",
        "hello",
        "Was it a car or a cat I saw?",
        "No 'x' in Nixon"
    ]

    for string in test_strings:
        result = is_palindrome(string)
        print(f"'{string}' is a palindrome: {result}")
```
